```python
import glob, os
from sklearn.feature_extraction.text import TfidfVectorizer

# Check working directory
print("CWD:", os.getcwd())

# List concept files
paths = glob.glob("math/train-medium/*.txt")
print("Found concept files:", paths)

# Load into docs list
docs = []
for path in paths:
    with open(path, encoding='utf-8') as f:
        docs.append(f.read())

print("Loaded docs count:", len(docs))

# Preview first document
if docs:
    print("Preview:", docs[0][:200], "…")

# Build TF-IDF model
concept_vec = TfidfVectorizer(stop_words='english', max_features=10000)
X_concepts = concept_vec.fit_transform(docs)
print("Concept matrix shape:", X_concepts.shape)
```
